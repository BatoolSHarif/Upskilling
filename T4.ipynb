```python
#Mission Two
#Q1
import csv
import re
file="C:\\Users\\Batool\\Desktop\\Upskilling-Python\\1 - Basics\\Part6 - REGEX\\T4 - After M8\\short_movies.csv"
with open(file,'r') as csvfile:
    reader=csv.reader(csvfile)
    matches=[]
    count=0
    count2=0
    count3=0
    reg=re.compile("good")
    reg2=re.compile("fun|funny")
    reg3=re.compile("great")
    for raw in reader:
        matches+= reg.findall(raw[4])
        matches+=reg2.findall(raw[4])
        matches+=reg3.findall(raw[4])
    for i in range (len(matches)):
        if matches[i]=="good":
            count=count+1
        elif matches[i]=="fun"or matches[i]=="funny":
            count2=count2+1
        elif matches[i]=="great":
            count3=count3+1
            
print("The \'good\' apperd for = ",count,"\nThe \'fun|funny' apperd for = ",count2,"\nThe \'great' apperd for = ",count3)
```
